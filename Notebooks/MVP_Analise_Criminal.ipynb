## 🧠 Tipo de Problema

Este projeto caracteriza-se como uma análise exploratória de dados (EDA), com foco na compreensão de padrões, distribuições e relações entre variáveis da criminalidade.

Adicionalmente, realiza-se uma classificação descritiva dos estados em níveis de criminalidade (baixo, médio e alto), baseada na distribuição da taxa por 100 mil habitantes.

Essa classificação não utiliza algoritmos de aprendizado de máquina, sendo construída a partir de critérios estatísticos (quantis).


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# carregar base
df = pd.read_csv('sua_base.csv', sep=';', encoding='latin-1')

# ajuste de tipos
df['Ano'] = df['Ano'].astype(int)
df['Ocorrencias'] = df['Ocorrencias'].astype(float)
df['Populacao'] = df['Populacao'].astype(float)

# criação da taxa
df['TX_100MIL'] = (df['Ocorrencias'] / df['Populacao']) * 100000

df.head()
```

```python
# agregação correta por UF
df_uf = df.groupby('UF').agg({
    'Ocorrencias': 'sum',
    'Populacao': 'sum'
}).reset_index()

df_uf['TX_100MIL'] = (df_uf['Ocorrencias'] / df_uf['Populacao']) * 100000

df_uf.head()
```

```python
# classificação por quantis
df_uf['nivel_crime'] = pd.qcut(
    df_uf['TX_100MIL'],
    q=3,
    labels=['Baixo', 'Médio', 'Alto']
)

df_uf
```

```python
# gráfico distribuição
sns.countplot(x='nivel_crime', data=df_uf)
plt.title('Distribuição dos Estados por Nível de Criminalidade')
plt.show()
```

```python
# boxplot
sns.boxplot(x='nivel_crime', y='TX_100MIL', data=df_uf)
plt.title('Distribuição da Taxa por Nível de Criminalidade')
plt.show()
```
