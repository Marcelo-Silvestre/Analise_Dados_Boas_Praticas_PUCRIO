## 📌 Descrição do Problema
Analisar a criminalidade no Brasil considerando diferenças entre estados, anos e tipos de crime, utilizando taxa por 100 mil habitantes para comparações adequadas.

## 🎯 Hipóteses
1. Estados mais populosos não têm necessariamente maiores taxas

3. Há tendências temporais
4. Existem diferenças regionais

EDA com classificação descritiva baseada em quantis, sem uso de ML.

## 🔍 Estrutura dos Dados

```python
df.info()df.describe()
```

## 📊 Crimes por Tipo

```python
df.groupby('Tipo Crime')['Ocorrencias'].sum().sort_values(ascending=False).plot(kind='bar')plt.title('Crimes por Tipo')plt.show()
```

## 📈 Evolução Temporal

```python
df_group = df.groupby(['Ano','Tipo Crime'])['Ocorrencias'].sum().reset_index()sns.lineplot(data=df_group, x='Ano', y='Ocorrencias', hue='Tipo Crime')plt.show()
```

## ⚖️ Taxa por UF (correta)

```python
df_uf = df.groupby('UF').agg({'Ocorrencias':'sum','Populacao':'sum'}).reset_index()df_uf['TX_100MIL'] = (df_uf['Ocorrencias']/df_uf['Populacao'])*100000df_uf.sort_values('TX_100MIL').plot(x='UF', y='TX_100MIL', kind='bar')plt.show()
```

## 🏆 Top 10

```python
top10 = df_uf.sort_values('TX_100MIL', ascending=False).head(10)top10.plot(x='UF', y='TX_100MIL', kind='bar')plt.show()
```

## 📦 Boxplot

```python
sns.boxplot(x='Tipo Crime', y='TX_100MIL', data=df)plt.xticks(rotation=45)plt.show()
```

## 🔥 Correlação

```python
corr = df[['Ocorrencias','Populacao','TX_100MIL']].corr()sns.heatmap(corr, annot=True)plt.show()
```

## ⚙️ Classificação Descritiva

```python
df_uf['nivel_crime'] = pd.qcut(df_uf['TX_100MIL'], q=3, labels=['Baixo','Médio','Alto'])sns.countplot(x='nivel_crime', data=df_uf)plt.show()
```

## 📈 Análise Temporal por UF

```python
df_uf_ano = df.groupby(['UF','Ano']).agg({'Ocorrencias':'sum','Populacao':'sum'}).reset_index()df_uf_ano['TX_100MIL'] = (df_uf_ano['Ocorrencias']/df_uf_ano['Populacao'])*100000sns.lineplot(data=df_uf_ano, x='Ano', y='TX_100MIL', hue='UF', legend=False)plt.show()
```
